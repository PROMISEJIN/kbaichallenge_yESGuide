```python
# 특성추출에서 기업 추출
from tqdm import tqdm
import pandas as pd

df_final_tone11 = pd.read_csv('./newdata_ESG기사3.csv')
# df_final_tone21 = pd.read_csv('./df_final_tone21.csv')
# df_final_tone = pd.concat([df_final_tone11,df_final_tone21])

data = df_final_tone11[['특성추출(가중치순 상위 50개)','label', 'score', 'label.1', 'score.1', 'label.2', 'score.2']].reset_index(drop = True)

# 기업명 가져오기
ESGlist = pd.read_csv('./ESGlist.csv', encoding='cp949')
stock_list = ESGlist['기업명'].to_list()

# 제목에서 종목명 추출

data['종목명'] = 'default'
data_mt = []    # multiple title

for sl in tqdm(stock_list) :
    for i in range(len(data)) :
        if sl in data.loc[i, '특성추출(가중치순 상위 50개)'].split(',') :       # 특성추출에 종목명이 있을 때(대문자로 통일)
            if data.loc[i, '종목명'] == 'default' :    # 종목명이 하나인 경우
                data.loc[i, '종목명'] = sl             # 종목명 부여
                data_d = data.loc[i]
                data_mt.append(data_d)
            else :                                     # 종목명이 여러개인 경우
                row_tmp = data.loc[i]                  # 행 복사
                row_tmp['종목명'] = sl                 # 종목명 부여
                data_mt.append(row_tmp)
data_mt = pd.DataFrame(data_mt)
data_mt
```
