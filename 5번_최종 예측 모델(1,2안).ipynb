#### 1-1 모델

```python
## K-fold 사용

from random import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, StratifiedKFold

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

#실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)

kfold = KFold(n_splits= 5, shuffle= True, random_state=3)
skf = StratifiedKFold(n_splits=5)

#빈 accuracy 배열
acc_score = []

# 데이터 분류

# # 원핫인코딩해주기
# y_encoded = to_categorical(train_set['ESG등급_y'])

def model2_fn():
    model2 = Sequential()
    model2.add(Dense(36, input_dim = 3, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    model2.add(Dense(48, activation = 'relu'))
    model2.add(Dense(6,activation = 'softmax'))
    return model2

for train_index, test_index in skf.split(yes_Three_mean_total.iloc[:,1:4],yes_Three_mean_total.iloc[:,4:5]):
    x_train, x_test = yes_Three_mean_total.iloc[:,1:4], yes_Three_mean_total.iloc[:,1:4]
    y_train, y_test = yes_Three_mean_total.iloc[:,4:5], yes_Three_mean_total.iloc[:,4:5]
    
    model2 = model2_fn()
    
    #모델 컴파일
    model2.compile(loss='sparse_categorical_crossentropy',
                optimizer = 'adam',
                metrics=['accuracy'])


    # 모델 실행
    history = model2.fit(x_train, y_train, epochs=200, batch_size = 10)

    #결과 출력
    accuracy = model2.evaluate(x_test,y_test)[1]
    acc_score.append(accuracy)
```

```python
avg_acc_score = sum(acc_score) / 5
print('정확도: ', acc_score)
print('정확도 평균',avg_acc_score)
```

```python
ESGlist_2021 = pd.read_csv('./ESGlist.csv', encoding='cp949')
ESGlist_2021.drop(['Unnamed: 0','기업코드','환경','사회','지배구조','비고'], axis=1, inplace=True)
ESGlist_2021.columns = ESGlist_2021.columns.str.replace('기업명','종목명')
ESGlist_2021['ESG등급'] = ESGlist_2021['ESG등급'].replace(['A+', 'A' , 'B+', 'B','C','D'], [3,2,2,1,1,0])
yes_Three_mean_total=pd.merge(yes_Three_mean_total, ESGlist_2021 , on = '종목명' , how ='left')
yes_Three_mean_total = yes_Three_mean_total.dropna()
yes_Three_mean_total
```

```python
## K-fold 사용

from random import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, StratifiedKFold

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

#실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)

kfold = KFold(n_splits= 5, shuffle= True, random_state=3)
skf = StratifiedKFold(n_splits=5)

#빈 accuracy 배열
acc_score = []

# 데이터 분류

# # 원핫인코딩해주기
# y_encoded = to_categorical(train_set['ESG등급_y'])

def model2_fn():
    model2 = Sequential()
    model2.add(Dense(36, input_dim = 3, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    model2.add(Dense(48, activation = 'relu'))
    model2.add(Dense(6,activation = 'softmax'))
    return model2

for train_index, test_index in skf.split(yes_Three_mean_total.iloc[:,1:4],yes_Three_mean_total.iloc[:,4:5]):
    x_train, x_test = yes_Three_mean_total.iloc[:,1:4], yes_Three_mean_total.iloc[:,1:4]
    y_train, y_test = yes_Three_mean_total.iloc[:,4:5], yes_Three_mean_total.iloc[:,4:5]
    
    model2 = model2_fn()
    
    #모델 컴파일
    model2.compile(loss='sparse_categorical_crossentropy',
                optimizer = 'adam',
                metrics=['accuracy'])


    # 모델 실행
    history = model2.fit(x_train, y_train, epochs=200, batch_size = 10)

    #결과 출력
    accuracy = model2.evaluate(x_test,y_test)[1]
    acc_score.append(accuracy)
```

```python
avg_acc_score = sum(acc_score) / 5
print('정확도: ', acc_score)
print('정확도 평균',avg_acc_score)
```

```python
## K-fold 사용

from random import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, StratifiedKFold

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

#실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)

kfold = KFold(n_splits= 5, shuffle= True, random_state=3)
skf = StratifiedKFold(n_splits=5)

#빈 accuracy 배열
acc_score = []

# 데이터 분류

# # 원핫인코딩해주기
# y_encoded = to_categorical(train_set['ESG등급_y'])

def model_fn():
    model = Sequential()
    model.add(Dense(36, input_dim = 3, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    # model.add(Dense(48, activation = 'relu'))
    model.add(Dense(48, activation = 'relu'))
    model.add(Dense(6,activation = 'softmax'))
    return model

for train_index, test_index in skf.split(train_set.iloc[:,1:4],train_set.iloc[:,4:5]):
    x_train, x_test = train_set.iloc[:,1:4], train_set.iloc[:,1:4]
    y_train, y_test = train_set.iloc[:,4:5], train_set.iloc[:,4:5]
    
    model = model_fn()
    
    #모델 컴파일
    model.compile(loss='sparse_categorical_crossentropy',
                optimizer = 'adam',
                metrics=['accuracy'])


    # 모델 실행
    history = model.fit(x_train, y_train, epochs=200, batch_size = 10)

    #결과 출력
    accuracy = model.evaluate(x_test,y_test)[1]
    acc_score.append(accuracy)
```

```python
avg_acc_score = sum(acc_score) / 5
print('정확도: ', acc_score)
print('정확도 평균',avg_acc_score)
```
